```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
df = pd.read_csv('dataset/agff_cleaned.csv')
```

```python
df.head()
```

```python
df.info()
```

### Performing EDA

```python
df_copy = df.drop(['day', 'month', 'year'], axis=1) ### We can drop the date for not
```

```python
df_copy['Classes'].value_counts() ### Lets see the value counts
```

- There are a lot of classes that are similar, probably because of empty spaces
- We can strip these strings to get required string

```python
df_copy['Classes'] = np.where(df_copy['Classes'].str.contains("not fire"), 0, 1) ### Removes trailing spaces
```

```python
df_copy.Classes.value_counts()
```

```python
plt.style.use('ggplot')
```

```python
df_copy.hist(bins=50, figsize=(20, 15))
plt.show() ### Lets plot histograms for all the features
```

```python
percentage = df_copy.Classes.value_counts(normalize=True) * 100
```

```python
classlabels = ["Fire", "Not Fire"] ### Pi chart to visualize fire and not fire
plt.figure(figsize=(12, 7))
plt.pie(percentage, labels=classlabels, autopct='%1.1f%%')
plt.plot()
```

```python
df_copy.corr() ### Finding the correlation between all features
```

```python
sns.heatmap(df_copy.corr()) ### Plotting a heatmap of the correlation plot
```

```python
sns.boxplot(df_copy['FWI']) ### Box plot for FWI
```

```python
sns.boxenplot(df_copy['FWI'])
```

```python
df['Classes'] = np.where(df['Classes'].str.contains("not fire"), 0, 1)
```

```python
### monthly fire analysis for region 1
dftemp = df.loc[df['Region'] == 1]
plt.subplots(figsize=(12, 6))
sns.set_style('whitegrid')
sns.countplot(x='month', hue='Classes', data=df)
plt.xlabel("Months")
plt.ylabel("Number of fires")
plt.title("Fires in Sidi-Bel Abbes region")
plt.plot()
```

```python
df.Region.value_counts()
```

```python
### monthly fire analysis for region 0
dftemp = df.loc[df['Region'] == 0]
plt.subplots(figsize=(12, 6))
sns.set_style('whitegrid')
sns.countplot(x='month', hue='Classes', data=df)
plt.xlabel("Months")
plt.ylabel("Number of fires")
plt.title("Fires in Bejaia region")
plt.plot()
```
